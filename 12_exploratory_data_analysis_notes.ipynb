### <a name="mark_01"></a>Análisis Exploratorio de Datos?
[Index](#index)


Organizar y entender las variables: podrás identificar los diferentes tipos de variables, las categorías a la que pertenecen y el tipo de análisis que puedes realizar sobre ellas.

Establecer relaciones entre las variables

Encontrar patrones ocultos en los datos: podrás encontrar información o comportamientos relevantes cuando hagas el EDA.

Ayuda a escoger el modelo correcto para la necesidad correcta: una vez encuentres como están relacionadas las variables podrás descubrir las variables que mas se ajustan a un tipo de modelo y de esta manera eligiras el correcto

Ayuda a tomar decisiones informadas: decisiones basadas en los datos, en las relaciones que encuentres entre variables, en patrones ocultos y en los modelos que generes a través de la EDA

Pasos de una Análisis Exploratorio de Datos

Hacer preguntas sobe los datos. Hazte las siguientes preguntas para guiar el EDA:

- Que te gustaria encontrar?
- Que quisieras saber de los datos?
- Cual es la razon para realizar el analisis?

Determinar el tamaño de los datos. Debes responder preguntas como:

- Cuantas observaciones existen?
- Cuantas variables hay?
- Necesito todas las observaciones?
- Necesito todas las variables?

Categorizar las variables. Debes preguntarte:

- Cuantas variables categóricas existen?
- Cuantas variables continuas existen?
- Como puedo explorar cada variable dependiendo de su categoría?

Limpieza y validación de los datos. En ese paso debes preguntarte:

- Tengo valores faltantes?
- Cual es la proporción de datos faltantes?
- Como puedo tratar a los datos faltantes?
- Cual es la distribución de los datos?
- Tengo valores atipicos?

Establecer relaciones entre los datos. Responde preguntas como:

- Existe algun tipo de relacion entre mi variable X y Y?
- Que pasa ahora si considero la variable Z en el analisis?
- Que significa que las observaciones se agrupen?
- Que significa el patron que se observa?

**Este proceso es ciclico. A pesar de que pueda parecer infinito, este proceso en algun momento debe salir del ciclo y continuar para obtener algun tipo de valor**

### <a name="mark_02"></a>Tipos de Análisis de Datos
[Index](#index)

Tipos de Analisis de Datos

Se dividen en 4 partes:

* Analítica Descriptiva: Que sucedió? Provee de ideas sobre eventos del pasado

* Analítica Diagnostica: Porque sucedió? Profundiza para encontrar las causas del evento

* Analítica Predictiva: Que podría pasar si? Utiliza los datos del pasado para predecir un futuro evento

* Analítica Prescriptiva: Que debería hacerse? Analiza decisiones y eventos del pasado para estimar la probabilidad de diferentes resultados.

Todas las analíticas se van construyendo una sobre la otra por lo que lleva tiempo, es decir, no puedes pasar de la Descriptiva a la Predictiva sin antes pasar por la Diagnostica

A medida que vas pasando de una analitica a otra tambien incrementa la complejidad ya que necesitas entender los datos a un nivel mas preciso conforme vayas avanzando

### <a name="mark_24"></a>Funcion de Distribucion Acumulada (CDFs)
[Index](#index)

Devuelve la probabilidad de que una variable sea igual o menor que un valor determinado. 

Ejemplo, la probabilidad de que en un salon hayan personas menores o iguales a una edad especifica.

```python
#Usando seaborn
sns.ecdfplot(
    data=df_clean_null,
    x='flipper_length_mm'
)
```

```python
#usando empiricaldist:
df_cdf=empiricaldist.Cdf.from_seq(
    df_clean_null.flipper_length_mm,
    normalize=True
)
df_cdf
```

```python
#Encontrar un pinguino que tenga una long de ala de 200 o menos:
df_cdf.plot()

length=200
prob=df_cdf.forward(length)

print(f'longitud: {length}\nProbabilidad: {prob}')

plt.vlines(
    x=length, #El valor de "x"
    ymin=0, #El min valor de "y" para graficar la línea vertical
    ymax=prob, #El max valor de "y" para graficar la línea vertical
    color='black',
    linestyle='dashed'
)

plt.hlines(
    y=prob,
    xmin=df_cdf.qs[0],#qs[0] es el primer valor de la primer columna longitud
    xmax=length,
    color='black',
    linestyle='dashed'
)

plt.plot(length, prob, 'ro')
```

```python
#Prueba para ver los nombres de las columnas, pero los perdío en la transformación
df_cdf_pandas=pd.DataFrame(df_cdf)

df_cdf_pandas.columns

df_cdf_pandas
```

```python
df_cdf.step ( )# Es el gráfico acumulativo

#Condiciones de probabilidad que queremos averiguar:
ps = ( 0.25 , 0.75 ) # en este caso es el IQR pero puede ser cualquiera 

#Si preguntabamos sobre la longitud para obtener la probabilidad, ahora la invertimos.
qs = df_cdf.inverse(ps)

#parametros para las líneas de intersección.
plt.vlines (
    x = qs,
    ymin = 0,
    ymax=ps,
    color='black',
    linestyle='dashed'
)

plt.hlines(
   y=ps,
   xmin=df_cdf.qs[0],
   xmax=qs,
   color='black',
   linestyle='dashed'
)

plt.scatter(
   x=qs,
   y=ps,
   color='red',
   zorder=2
)
```

### <a name="mark_26"></a>Ley De Los Grandes Números:
[Index](#index)

**"Cuanto más tiradas más nos acercamos a la probabilidad teórica"**

```python
dice = empiricaldist.Pmf.from_seq([1, 2, 3, 4, 5, 6])#Un dado de 6 caras iguales
dice.bar()
```

```python
#Tirando el dado n veces:
for sample_size in (1e2, 1e3, 1e4): #100, 1000, 10.000 tiradas
    sample_size=int(sample_size) #la función sample() recibe integers.
    #genera una lista del 1 al 6 del objeto "dice" aleatoreament para cada sample_size:
    #primero, genera una lista aleatorea del 1 al 6 con 100 elementos
    #segundo, genera una lista aleatorea del 1 al 6 con 1000 elementos
    #tercero, genera una lista aleatorea del 1 al 6 con 10.000 elementos
    values=dice.sample(sample_size)
    #aplicamos la función Pmf a nuestra sequencia de valores
    sample_pmf=empiricaldist.Pmf.from_seq(values)
    
    #graficando cada caso
    plt.figure(figsize=(5,5))
    sample_pmf.bar()
    plt.axhline(y=1/6, color='red', linestyle='dashed')
    plt.ylim([0,0.3])
    plt.title(f'Sample size: {sample_size}')
```

### <a name="mark_31"></a>Limitante Del Coeficiente De Correlación Lineal:
[Index](#index)

```python
#Correlación lineal en una cuadrática.
x=np.linspace(-100,100,100)
y=x**2
y += np.random.normal(10,1000, x.size)

sns.scatterplot(
    x=x, y=y
)

#correlación lineal usando NumPy
np.corrcoef(x, y)#No se observa correlación
```
